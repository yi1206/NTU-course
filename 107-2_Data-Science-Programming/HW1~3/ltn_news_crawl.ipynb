```python
import os
import requests
from bs4 import BeautifulSoup
from google import google

# Google search api
num_page = 1

#  site:news.ltn.com.tw/news   自由時報

ltn_news_links = []
ltn_news = []
media = [" site:news.ltn.com.tw/news"]
word = ["毒品販運","詐欺","走私","稅務犯罪","組織犯罪","證券犯罪","貪污賄賂","第三方洗錢"]

for i in media :
    for j in word:
        search_res = google.search(j+i,num_page);
        for m in search_res:
            ltn_news_links.append(m.link)
#             print(m.link)
print("自由時報 : " + str(len(ltn_news_links)) + "筆" + "相關新聞" )
print(ltn_news_links)
print("-----------------------------------------------------------------------")


for link in ltn_news_links :
    if (link == None):
        continue
    r = requests.get(link).text
    soup = BeautifulSoup(r,"html.parser")
    soup = soup.find( "div" , "text" )
    soup = soup.find('p')
    if (soup == None):
        continue
#     print(soup.get_text())
    ltn_news.append( soup.get_text())

print("自由時報 相關新聞內文 ")
print(ltn_news)
```
