```python
import csv
from io import StringIO
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs

#用來抓資料開始與結束年分的function
def yearzone(doc):
    nodes = doc.select('option')
    startyear=10000
    endyear=0
    
    for n in nodes:
        year = n.get_text()
        year=int(year)
        if startyear > year:
            startyear = year
        if endyear < year:
            endyear = year
    return [startyear,endyear]
```

```python
#用requests和bs4抓資料開始與結束年分
res=requests.get('https://beta.bls.gov/dataViewer/view/timeseries/CES0000000001')
doc = bs(res.text, 'lxml')
year = yearzone(doc)
print(year)
```

```python
#用requests.post請求所需資料
url = 'https://beta.bls.gov/dataViewer/csv'
urlres=requests.post(url, data={'selectedSeriesIds': 'CES0000000001', 'startYear': year[0], 'endYear': year[1]})
#取出來存成dataframe的資料格式
data=urlres.text
data_io = StringIO(data)
df =  pd.read_csv(data_io)

df.head(20)    
```
