```python
#----------資料讀入並進行篩選----------
import pandas as pd

# 讀檔
TechnologyEquityETF= pd.read_csv('Technology Equity ETF List (75).csv')
# 根據Inception欄位進行篩選
fliter = (TechnologyEquityETF["Inception"] < '2015-12-31') 
newTechnologyEquityETF = TechnologyEquityETF[fliter] 

#將符合的ETF代碼轉換為List
TechnologyEquityETFList = newTechnologyEquityETF['Symbol'].values.tolist()
 
print(TechnologyEquityETFList)
```

```python
#設置日期範圍
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import calendar
start_date = "2015-12-31"
stop_date = "2019-03-22"
start = datetime.strptime(start_date, "%Y-%m-%d")
stop = datetime.strptime(stop_date, "%Y-%m-%d")
print(start)
print(stop)
```

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
import datetime
def process_etf_str(etf_str, Symbol,all_data_table):
    
#    print("function_etf_str:",end="")
#    print(etf_str)
    
    etf_str_list = etf_str.split(' ')
#    print("etf_str_list:",end="")
#    print(etf_str_list)
    
    time = etf_str_list[0].split(',')
#    print("time:",end="")
#    print(time)
    
    etf_close_price = etf_str_list[2].split(',')
#    print("etf_close_price:",end="")
#    print(etf_close_price)
    
    dic_etf_close_price = {
        'Date' : time,
        Symbol : etf_close_price
    }
    
    tmp_df = pd.DataFrame(dic_etf_close_price)
#    print(tmp_df)
    if all_data_table.empty:
#        print("alldatatable為空 指派")
        all_data_table = tmp_df
    else:
#        print("進入else 準備join")
        all_data_table = all_data_table.merge(tmp_df, how='outer')
#        all_data_table = all_data_table.set_index('time').join(tmp_df.set_index('time'))
#        print("after join:")
#        print(all_data_table)
        
    return all_data_table
```

```python
all_data_table = pd.DataFrame()
#count = 0
for Symbol in TechnologyEquityETFList:
#    print('start crawling :', Symbol)
    res = requests.get('https://www.moneydj.com/etf/x/Basic/ETF/X/xdjbcd/Basic0003BCD.xdjbcd?etfid='+Symbol+'&b='+start_date+'&c='+stop_date)
#    print(res.text)
    doc = bs(res.text, 'lxml')
    if(doc):
        etf_str = doc.p.string #選擇p裡面的string
#    print("etf_str:",end="")
#    print(etf_str)
    if len(etf_str):
        all_data_table = process_etf_str(etf_str,Symbol,all_data_table)
#    count += 1
#    if count == 5:
#        break
print(all_data_table.sort_values(by=['Date']))       
```
